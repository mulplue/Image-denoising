```python
import numpy as np
def get_noisy_mask(ratio, shape):
    mask = []
    H, W = shape[0], shape[1]
    for row in range(H):
        line = np.ones(W)
        line[:int(W*ratio)] = 0
        np.random.shuffle(line)
        mask.append(line)
    return np.array(mask)
```
